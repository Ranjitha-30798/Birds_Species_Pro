```python
#DF2 GRASS LAND DATA CLEANING PROCESS
```

```python
df11.drop(columns='Sub_Unit_Code', inplace=True)
```

```python
cat_cols = [
    'Admin_Unit_Code', 'Plot_Name', 'Location_Type', 'Observer',
    'Interval_Length', 'ID_Method', 'Distance', 'Sky', 'Wind',
    'Disturbance', 'Sex', 'Common_Name', 'Scientific_Name', 'AOU_Code'
]
df11[cat_cols] = df11[cat_cols].astype('category')
```

```python
df11['ID_Method'] = df11['ID_Method'].fillna(df11['ID_Method'].mode()[0])
```

```python
df11['AcceptedTSN'] = df11['AcceptedTSN'].fillna(-1).astype('Int64')
df11['TaxonCode'] = df11['TaxonCode'].fillna(-1).astype('Int64')
```

```python
df11['Date'] = pd.to_datetime(df11['Date'], errors='coerce')
```

```python
df11['Start_Time'] = pd.to_datetime(df11['Start_Time'], format='%H:%M', errors='coerce').dt.time
df11['End_Time'] = pd.to_datetime(df11['End_Time'], format='%H:%M', errors='coerce').dt.time
```

```python
df11['Start_Time'].fillna(pd.to_datetime('00:00').time(), inplace=True)
df11['End_Time'].fillna(pd.to_datetime('00:00').time(), inplace=True)
```

```python
from datetime import datetime, date, timedelta


df11['Start_Time'] = df11['Start_Time'].apply(lambda t: datetime.combine(date.today(), t))
```

```python
from datetime import datetime, date

df11['End_Time'] = df11['End_Time'].apply(lambda t: datetime.combine(date.today(), t))

```

```python
print(df11['Start_Time'].apply(type).unique())
print(df11['End_Time'].apply(type).unique())
```

```python
distance_map1 = {
    '<= 50 Meters': 25,
    '50 - 100 Meters': 75,
    '100 - 200 Meters': 150,
    '200+ Meters': 250
}
df11['distance_numeric'] = df11['Distance'].map(distance_map1)
```

```python
df11 = df11.dropna(subset=['distance_numeric'])
```

```python
df11['Distance'] = pd.to_numeric(df11['distance_numeric'], errors='coerce')


df11['Distance'] = df11['Distance'].fillna(df11['Distance'].mean())
```

```python
df11['Interval_Minutes'] = df11['Interval_Length'].str.extract(r'(\d+)').astype(float)
```

```python
if 'Unknown' not in df11['Sex'].cat.categories:
    df11['Sex'] = df11['Sex'].cat.add_categories(['Unknown'])


df11['Sex'].fillna('Unknown', inplace=True)

```

```python
bool_cols = [
    'Flyover_Observed', 'PIF_Watchlist_Status', 'Regional_Stewardship_Status',
    'Previously_Obs', 'Initial_Three_Min_Cnt'
]
df11[bool_cols] = df11[bool_cols].astype(bool)
```

```python
#1. Temporal Analysis
```

```python
#Seasonal Trends:
df11['Month'] = df11['Date'].dt.month
df11['Season'] = df11['Month'].map({12:'Winter', 1:'Winter', 2:'Winter',
                                  3:'Spring', 4:'Spring', 5:'Spring',
                                  6:'Summer', 7:'Summer', 8:'Summer',
                                  9:'Fall', 10:'Fall', 11:'Fall'})

species_by_season = df11.groupby('Season')['Scientific_Name'].nunique()
species_by_season.plot(kind='bar', title='Species Diversity by Season')
```

```python
#Observation Time:
df11['start_hour'] = pd.to_datetime(df11['Start_Time'], errors='coerce').dt.hour
df11['end_Hour'] = pd.to_datetime(df11['End_Time'], errors='coerce').dt.hour

if df11['start_hour'].notna().sum() > 0:
    df11['start_hour'].value_counts().sort_index().plot(kind='bar', title='Observations by Hour')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Observations')
    plt.tight_layout()
    plt.show()
else:
    print("No valid 'start_hour' values to plot.")
```
