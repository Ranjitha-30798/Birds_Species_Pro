```python
#1. Temporal Analysis
```

```python
# Seasonal Trends: Analyze the Date and Year columns to detect patterns in bird sightings across different seasons or years
df11['Month'] = df11['Date'].dt.month
df11['Season'] = df11['Month'].map({12:'Winter', 1:'Winter', 2:'Winter',
                                  3:'Spring', 4:'Spring', 5:'Spring',
                                  6:'Summer', 7:'Summer', 8:'Summer',
                                  9:'Fall', 10:'Fall', 11:'Fall'})

species_by_season = df11.groupby('Season')['Scientific_Name'].nunique()
species_by_season.plot(kind='bar', title='Species Diversity by Season')
```

```python
# Observation Time: Study the Start_Time and End_Time to determine if specific time windows correlate with higher bird activity.
df11['start_hour'] = pd.to_datetime(df11['Start_Time'], errors='coerce').dt.hour
df11['end_Hour'] = pd.to_datetime(df11['End_Time'], errors='coerce').dt.hour

if df11['start_hour'].notna().sum() > 0:
    df11['start_hour'].value_counts().sort_index().plot(kind='bar', title='Observations by Hour')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Observations')
    plt.tight_layout()
    plt.show()
else:
    print("No valid 'start_hour' values to plot.")
```

```python
#Spatial Analysis
```

```python
#Location Insights: Group data by Location_Type (e.g., Grassland) to identify biodiversity hotspots.
df11.groupby('Location_Type')['Scientific_Name'].nunique().plot(kind='bar', title='Species Diversity by Habitat')
```

```python
#Plot-Level Analysis: Compare observations across different Plot_Name to see which plots attract more species or specific kinds of birds.
df11.groupby('Plot_Name')['Scientific_Name'].nunique().nlargest(10).plot(kind='bar', title='Top 10 Plots by Species Count')
```

```python
#3 Species Analysis
```

```python
#Diversity Metrics: Count unique species (Scientific_Name) observed and their distribution across Location_Type.Diversity Metrics:
total_unique_species = df11['Scientific_Name'].nunique()
species_by_location = df11.groupby('Location_Type')['Scientific_Name'].nunique()
species_by_location.plot(kind='bar', title='Species Diversity by Location Type')
plt.xlabel('Location Type')
plt.ylabel('Number of Unique Species')
plt.tight_layout()
plt.show()
```

```python
#Activity Patterns: Check the Interval_Length and ID_Method columns to identify the most common activity types (e.g., Singing).
df11['ID_Method'].value_counts().plot(kind='bar', title='Identification Methods')
```

```python
# Sex Ratio: Analyze the Sex column to understand the male-to-female ratio for different species.
df_filtered = df11[df11['Sex'].isin(['Male', 'Female'])]
sex_counts = df_filtered.groupby(['Scientific_Name', 'Sex']).size().unstack(fill_value=0)
sex_counts['Male/Female Ratio'] = (sex_counts['Male'] / sex_counts['Female']).round(2)
top_species = df_filtered['Scientific_Name'].value_counts().head(5).index
subset = sex_counts.loc[top_species]
subset[['Male', 'Female']].plot(kind='bar', title='Sex Distribution by Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#4 Environmental Conditions
```

```python
# Weather Correlation: Explore how Temperature, Humidity, Sky, and Wind impact observations, such as the number of birds or their distances.
import seaborn as sns
sns.scatterplot(data=df11, x='Temperature', y='Humidity', hue='Location_Type')
```

```python
# Disturbance Effect: Assess the impact of Disturbance (e.g., slight effect) on bird sightings.
df11.groupby('Disturbance')['Scientific_Name'].nunique().plot(kind='bar', title='Species Count by Disturbance Level')
```

```python
#5 Distance and Behavior
```

```python
#Distance Analysis: Evaluate the Distance column to identify species typically observed closer or farther from the observer.
df11.groupby('Scientific_Name')['Distance'].median().sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
#Flyover Frequency: Examine the Flyover_Observed column to detect trends in bird behavior during observation.
df11['Flyover_Observed'].value_counts(normalize=True).plot(kind='pie', autopct='%1.1f%%', title='Flyover Frequency')
```

```python
#6 Observer Trends
```

```python
#Observer Bias: Analyze data by Observer to check if specific individuals report more observations or certain species.
df11['Observer'].value_counts().head(10).plot(kind='bar', title='Top Observers by Records')
```

```python
#Visit Patterns: Evaluate the Visit column to see how repeated visits affect species count or diversity.
df11.groupby('Visit')['Scientific_Name'].nunique().plot(kind='line', title='Species Diversity by Visit Number')
```

```python
# 7 Conservation Insights
```

```python
#Watchlist Trends: Use the PIF_Watchlist_Status and Regional_Stewardship_Status to identify trends in species that are at risk or require conservation focus.
df11['PIF_Watchlist_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='PIF Watchlist Status')
```

```python
#Regional Stewardship:
df11.groupby('Regional_Stewardship_Status')['Scientific_Name'].nunique().plot(kind='bar', title='Species by Stewardship Status')
```

```python
# AOU Code Patterns: Study the distribution of species based on their AOU_Code to correlate with regional or national conservation priorities.
aou_distribution = df11['AOU_Code'].value_counts().sort_index()
aou_distribution.plot(kind='line', title='Species Distribution by AOU Code', figsize=(10, 5))
plt.xlabel('AOU Code')
plt.ylabel('Number of Records')
plt.tight_layout()
plt.show()
```
